```python
import pandas as pd
```

```python
data = pd.read_csv('./datasets/airline-dataset/flights/flights.csv')
data.head()
```

```python
from scipy import stats
```

```python
stats.trim_mean(data['ArrDelay'], 0.1)
```

```python
data['ArrDelay'].mode()
```

```python
data['ArrDelay'].median()
```

```python
data['ArrDelay'].var()
```

```python
data['ArrDelay'].std()
```

```python
data['ArrDelay'].hist(bins=200)
```

```python
data['ArrDelay'].hist(bins=40)
```
